## рҹҚҺ 1. Principio de Equivalencia: La Gravedad como GeometrГӯa

---

### 1.1 El Principio de Equivalencia DГ©bil (Galileo-Newton)

Todos los cuerpos caen con la misma aceleraciГіn en un campo gravitatorio, independientemente de su masa o composiciГіn:

$$m_i \vec{a} = m_g \vec{g} \implies \vec{a} = \frac{m_g}{m_i}\vec{g}$$

La **igualdad de masa inercial y gravitatoria** ($m_i = m_g$) estГЎ verificada experimentalmente con precisiГіn $\sim 10^{-15}$ (experimento MICROSCOPE, 2017).

### 1.2 El Principio de Equivalencia de Einstein (Fuerte)

> **NingГәn experimento local puede distinguir entre un campo gravitatorio uniforme y un sistema de referencia uniformemente acelerado.**

Esto tiene consecuencias profundas:
- La gravedad **no es una fuerza** en el sentido convencional
- Es una manifestaciГіn de la **curvatura del espacio-tiempo**
- Las partГӯculas en caГӯda libre siguen **geodГ©sicas** (lГӯneas de extrema longitud)
- Siempre es posible encontrar coordenadas localmente inerciales (sistema en caГӯda libre)

### 1.3 Consecuencias Inmediatas

1. **Corrimiento gravitacional al rojo:** La luz pierde energГӯa al escapar de un pozo gravitacional:
$$\frac{\Delta\nu}{\nu} = -\frac{\Phi}{c^2}$$

2. **DeflexiГіn de la luz:** Los fotones siguen trayectorias curvas en un campo gravitatorio

3. **DilataciГіn temporal gravitacional:** Los relojes van mГЎs lentos en campos gravitatorios mГЎs intensos:
$$d\tau = \sqrt{1 + \frac{2\Phi}{c^2}} \, dt \approx \left(1 + \frac{\Phi}{c^2}\right) dt$$

### 1.4 De la Equivalencia a la GeometrГӯa

Si la gravedad es equivalente a la aceleraciГіn, y la aceleraciГіn es un efecto geomГ©trico (coordenadas no inerciales), entonces la **gravedad es geometrГӯa**. El campo gravitacional se codifica en la **mГ©trica del espacio-tiempo** $g_{\mu\nu}(x)$:

$$ds^2 = g_{\mu\nu}(x)\, dx^\mu dx^\nu$$

En ausencia de gravedad: $g_{\mu\nu} = \eta_{\mu\nu}$ (Minkowski). Con gravedad: $g_{\mu\nu}$ es una funciГіn no trivial de las coordenadas.

```python
# ============================================================================
# SecciГіn 1b: Principio de Equivalencia вҖ” VisualizaciГіn 3D
# ============================================================================

from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# --- 1. Curvatura del espacio-tiempo: Masa deformando una membrana 3D ---
fig = plt.figure(figsize=(16, 7))

ax1 = fig.add_subplot(121, projection='3d')

# Grilla del espacio
x_grid = np.linspace(-5, 5, 100)
y_grid = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_grid, y_grid)

# "Pozo" gravitacional: z = -M / sqrt(xВІ + yВІ + ОөВІ)
M_mass = 3.0
epsilon = 0.8  # regularizaciГіn para evitar singularidad
Z = -M_mass / np.sqrt(X**2 + Y**2 + epsilon**2)

# Superficie
surf = ax1.plot_surface(X, Y, Z, cmap='inferno', alpha=0.8, edgecolor='none',
                        rstride=2, cstride=2)

# Masa central
ax1.scatter([0], [0], [Z.min() - 0.3], color='yellow', s=200, zorder=5)
ax1.text(0, 0, Z.min() - 0.8, '$M$', fontsize=14, ha='center', color='yellow')

# GeodГ©sica (trayectoria curva)
t_geo = np.linspace(0, 2*np.pi, 200)
r_geo = 3.0
x_geo = r_geo * np.cos(t_geo)
y_geo = r_geo * np.sin(t_geo)
z_geo = -M_mass / np.sqrt(x_geo**2 + y_geo**2 + epsilon**2)
ax1.plot(x_geo, y_geo, z_geo, 'cyan', linewidth=3, label='GeodГ©sica (Гіrbita)')

ax1.set_title('Espacio-tiempo curvado por masa\n"La gravedad es geometrГӯa"', fontsize=13)
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_zlabel('Curvatura')
ax1.legend(fontsize=10)
ax1.view_init(elev=35, azim=45)

# --- 2. ComparaciГіn: Gravedad vs AceleraciГіn ---
ax2 = fig.add_subplot(122)

# Ascensor en campo gravitacional
rect_grav = plt.Rectangle((0.5, 0.5), 2, 3, fill=False, edgecolor='blue', linewidth=3)
ax2.add_patch(rect_grav)
ax2.annotate('', xy=(1.5, 0.2), xytext=(1.5, -0.5),
            arrowprops=dict(arrowstyle='->', color='red', lw=3))
ax2.text(1.5, -0.8, '$\\vec{g}$', fontsize=16, ha='center', color='red')

# Persona y objetos cayendo igual
ax2.plot(1.5, 2.0, 'o', markersize=20, color='green')  # cabeza
ax2.plot([1.5, 1.5], [1.0, 2.0], 'g-', linewidth=3)     # cuerpo
ax2.plot([1.3, 1.7], [1.0, 1.0], 'g-', linewidth=3)     # piernas

# Pelota cayendo
for y_ball in [3.0, 2.5, 2.0, 1.5]:
    alpha = 0.3 + 0.7 * (3.0 - y_ball) / 2.0
    ax2.plot(2.2, y_ball, 'o', markersize=10, color='orange', alpha=alpha)
ax2.annotate('', xy=(2.2, 1.3), xytext=(2.2, 3.2),
            arrowprops=dict(arrowstyle='->', color='orange', lw=2))

ax2.text(1.5, 4.0, 'Campo gravitatorio $g$', fontsize=13, ha='center', color='blue',
        bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.8))

# Cohete acelerando (derecha)
rect_acc = plt.Rectangle((4.5, 0.5), 2, 3, fill=False, edgecolor='red', linewidth=3)
ax2.add_patch(rect_acc)
ax2.annotate('', xy=(5.5, 4.5), xytext=(5.5, 3.8),
            arrowprops=dict(arrowstyle='->', color='red', lw=3))
ax2.text(5.5, 4.7, '$\\vec{a} = \\vec{g}$', fontsize=16, ha='center', color='red')

# Llama del cohete
from matplotlib.patches import Polygon
flame = Polygon([[5.0, 0.5], [6.0, 0.5], [5.5, -0.3]], color='orange', alpha=0.7)
ax2.add_patch(flame)

# Misma persona dentro
ax2.plot(5.5, 2.0, 'o', markersize=20, color='green')
ax2.plot([5.5, 5.5], [1.0, 2.0], 'g-', linewidth=3)
ax2.plot([5.3, 5.7], [1.0, 1.0], 'g-', linewidth=3)

# Pelota "cayendo" (realmente el piso sube)
for y_ball in [3.0, 2.5, 2.0, 1.5]:
    alpha = 0.3 + 0.7 * (3.0 - y_ball) / 2.0
    ax2.plot(6.2, y_ball, 'o', markersize=10, color='orange', alpha=alpha)
ax2.annotate('', xy=(6.2, 1.3), xytext=(6.2, 3.2),
            arrowprops=dict(arrowstyle='->', color='orange', lw=2))

ax2.text(5.5, 4.0, 'Cohete acelerando $a = g$', fontsize=13, ha='center', color='red',
        bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8))

# Signo de igualdad
ax2.text(3.5, 2.0, '$\\equiv$', fontsize=40, ha='center', va='center', color='purple')
ax2.text(3.5, 0.0, 'LOCALMENTE\nINDISTINGUIBLES', fontsize=11, ha='center', 
        color='purple', fontweight='bold')

ax2.set_xlim(-0.5, 8)
ax2.set_ylim(-1.5, 5.5)
ax2.set_aspect('equal')
ax2.axis('off')
ax2.set_title('Principio de Equivalencia de Einstein', fontsize=14)

plt.tight_layout()
plt.show()

# --- 3. AnimaciГіn: Curvatura del espacio-tiempo con masa variable ---
fig_3d, ax_3d = plt.subplots(subplot_kw={'projection': '3d'}, figsize=(10, 8))

def animate_curvature(frame):
    ax_3d.clear()
    M_anim = 0.1 + frame * 0.08  # Masa creciente
    Z_anim = -M_anim / np.sqrt(X**2 + Y**2 + epsilon**2)
    
    ax_3d.plot_surface(X, Y, Z_anim, cmap='magma', alpha=0.8, edgecolor='none',
                       rstride=3, cstride=3)
    
    # GeodГ©sica circular
    r_orb = max(2.0, 5.0 - frame * 0.05)
    x_orb = r_orb * np.cos(t_geo)
    y_orb = r_orb * np.sin(t_geo)
    z_orb = -M_anim / np.sqrt(x_orb**2 + y_orb**2 + epsilon**2)
    ax_3d.plot(x_orb, y_orb, z_orb, 'cyan', linewidth=2.5)
    
    ax_3d.set_zlim(-6, 0)
    ax_3d.set_xlim(-5, 5)
    ax_3d.set_ylim(-5, 5)
    ax_3d.set_title(f'Curvatura del espacio-tiempo вҖ” $M = {M_anim:.1f}$', fontsize=14)
    ax_3d.view_init(elev=35, azim=45 + frame * 2)
    return []

anim_curv = FuncAnimation(fig_3d, animate_curvature, frames=50, interval=120, blit=True)
plt.close(fig_3d)
HTML(anim_curv.to_jshtml())
```

## рҹ§® 2. GeometrГӯa Diferencial: El Lenguaje de la Relatividad General

---

### 2.1 Variedades Diferenciables

El espacio-tiempo es una **variedad diferenciable** de dimensiГіn 4: un espacio que localmente se parece a $\mathbb{R}^4$ pero puede tener topologГӯa global no trivial. Cada punto se etiqueta con coordenadas $x^\mu = (x^0, x^1, x^2, x^3) = (ct, x, y, z)$.

### 2.2 Tensor MГ©trico

La **mГ©trica** $g_{\mu\nu}$ es el objeto fundamental que codifica toda la geometrГӯa:

$$ds^2 = g_{\mu\nu}(x) \, dx^\mu \, dx^\nu$$

**Propiedades:**
- SimГ©trico: $g_{\mu\nu} = g_{\nu\mu}$ вҶ’ 10 componentes independientes en 4D
- Signatura $(-,+,+,+)$ (convenciГіn de Landau) o $(+,-,-,-)$
- $\det(g_{\mu\nu}) \neq 0$ (no degenerado)
- Define longitudes, ГЎngulos, volГәmenes y causalidad

**Ejemplos:**
- Minkowski: $g_{\mu\nu} = \eta_{\mu\nu} = \text{diag}(-1, 1, 1, 1)$
- Schwarzschild: $g_{00} = -(1-r_s/r)$, $g_{11} = (1-r_s/r)^{-1}$, $g_{22} = r^2$, $g_{33} = r^2\sin^2\theta$

### 2.3 ConexiГіn y Derivada Covariante

La **derivada covariante** $\nabla_\mu$ generaliza la derivada parcial a espacios curvos. ActГәa sobre vectores como:

$$\nabla_\mu V^\nu = \partial_\mu V^\nu + \Gamma^\nu_{\mu\lambda} V^\lambda$$

Los **sГӯmbolos de Christoffel** (conexiГіn de Levi-Civita) se calculan a partir de la mГ©trica:

$$\boxed{\Gamma^\lambda_{\mu\nu} = \frac{1}{2}g^{\lambda\sigma}\left(\partial_\mu g_{\nu\sigma} + \partial_\nu g_{\mu\sigma} - \partial_\sigma g_{\mu\nu}\right)}$$

**Propiedades:**
- SimГ©tricos: $\Gamma^\lambda_{\mu\nu} = \Gamma^\lambda_{\nu\mu}$ (torsiГіn nula)
- **No** son tensores (dependen de las coordenadas)
- Compatibles con la mГ©trica: $\nabla_\lambda g_{\mu\nu} = 0$
- En coordenadas cartesianas planas: $\Gamma^\lambda_{\mu\nu} = 0$

### 2.4 Transporte Paralelo

Un vector $V^\mu$ se **transporta paralelamente** a lo largo de una curva $x^\mu(\lambda)$ si:

$$\frac{DV^\mu}{D\lambda} \equiv \frac{dV^\mu}{d\lambda} + \Gamma^\mu_{\alpha\beta}\frac{dx^\alpha}{d\lambda}V^\beta = 0$$

En espacio plano, un vector transportado paralelamente no cambia. En espacio curvo, un vector transportado a lo largo de un circuito cerrado **puede rotar** вҶ’ esta rotaciГіn mide la curvatura.

## рҹӣӨпёҸ 3. GeodГ©sicas: El Movimiento Libre en Espacio-Tiempo Curvo

---

### 3.1 ВҝQuГ© es una GeodГ©sica?

Una **geodГ©sica** es la generalizaciГіn de una "lГӯnea recta" a un espacio curvo. Es la curva que **extremiza la longitud** (o el tiempo propio) entre dos puntos. En relatividad general, las partГӯculas libres (sin fuerzas no-gravitacionales) siguen geodГ©sicas.

### 3.2 EcuaciГіn GeodГ©sica

La ecuaciГіn de movimiento para una partГӯcula libre es:

$$\boxed{\frac{d^2 x^\mu}{d\tau^2} + \Gamma^\mu_{\alpha\beta}\frac{dx^\alpha}{d\tau}\frac{dx^\beta}{d\tau} = 0}$$

donde $\tau$ es un parГЎmetro afГӯn (tiempo propio para partГӯculas masivas).

**InterpretaciГіn:** El segundo tГ©rmino es la "fuerza gravitacional aparente" вҖ” pero en RG no es una fuerza real, sino el efecto de la **curvatura del espacio-tiempo**.

### 3.3 DerivaciГіn Variacional

Extremizando el funcional de longitud propia:

$$\delta \int d\tau = \delta \int \sqrt{-g_{\mu\nu} \frac{dx^\mu}{d\lambda}\frac{dx^\nu}{d\lambda}} \, d\lambda = 0$$

Las ecuaciones de Euler-Lagrange con $L = g_{\mu\nu}\dot{x}^\mu\dot{x}^\nu$ dan exactamente la ecuaciГіn geodГ©sica.

### 3.4 Tipos de GeodГ©sicas

- **Tipo tiempo** ($ds^2 < 0$): trayectorias de partГӯculas masivas. Extremizan el tiempo propio.
- **Tipo nulo** ($ds^2 = 0$): trayectorias de fotones (rayos de luz).
- **Tipo espacio** ($ds^2 > 0$): curvas de distancia mГӯnima en hipersuperficies espaciales.

### 3.5 DesviaciГіn GeodГ©sica

Dos geodГ©sicas inicialmente paralelas separadas por $\xi^\mu$ se desvГӯan segГәn la **ecuaciГіn de desviaciГіn geodГ©sica** (ecuaciГіn de Jacobi):

$$\frac{D^2 \xi^\mu}{D\tau^2} = -R^\mu_{\ \nu\alpha\beta} u^\nu u^\alpha \xi^\beta$$

donde $u^\mu = dx^\mu/d\tau$ es la 4-velocidad. **La curvatura causa convergencia/divergencia de geodГ©sicas cercanas** вҖ” esto es la esencia de las fuerzas de marea gravitacional.

## рҹ“җ 4. Curvatura: Del Tensor de Riemann al Escalar de Curvatura

---

### 4.1 ВҝCГіmo Medir la Curvatura?

La curvatura del espacio-tiempo se manifiesta de varias formas equivalentes:
1. **Los triГЎngulos no suman ПҖ** (exceso angular proporcional a la curvatura)
2. **Las geodГ©sicas inicialmente paralelas convergen o divergen** (desviaciГіn geodГ©sica)
3. **El transporte paralelo de un vector en un circuito cerrado lo rota** (holonomГӯa)

### 4.2 Tensor de Riemann

El **tensor de curvatura de Riemann** $R^\rho_{\ \sigma\mu\nu}$ mide la no-conmutatividad de la derivada covariante:

$$[\nabla_\mu, \nabla_\nu] V^\rho = R^\rho_{\ \sigma\mu\nu} V^\sigma$$

Se calcula a partir de la conexiГіn:

$$\boxed{R^\rho_{\ \sigma\mu\nu} = \partial_\mu \Gamma^\rho_{\nu\sigma} - \partial_\nu \Gamma^\rho_{\mu\sigma} + \Gamma^\rho_{\mu\lambda}\Gamma^\lambda_{\nu\sigma} - \Gamma^\rho_{\nu\lambda}\Gamma^\lambda_{\mu\sigma}}$$

**SimetrГӯas del tensor de Riemann:**
- $R_{\rho\sigma\mu\nu} = -R_{\sigma\rho\mu\nu} = -R_{\rho\sigma\nu\mu}$ (antisimГ©trico en pares)
- $R_{\rho\sigma\mu\nu} = R_{\mu\nu\rho\sigma}$ (simetrГӯa de intercambio)
- $R_{\rho[\sigma\mu\nu]} = 0$ (identidad algebraica de Bianchi)

En 4 dimensiones: **20 componentes independientes** (de las $4^4 = 256$ posibles).

### 4.3 Tensor de Ricci y Escalar de Curvatura

**Tensor de Ricci** (contracciГіn del tensor de Riemann):

$$R_{\mu\nu} = R^\lambda_{\ \mu\lambda\nu} = g^{\lambda\rho} R_{\rho\mu\lambda\nu}$$

Es simГ©trico: $R_{\mu\nu} = R_{\nu\mu}$ вҶ’ **10 componentes** en 4D.

**Escalar de curvatura** (traza del tensor de Ricci):

$$R = g^{\mu\nu} R_{\mu\nu}$$

### 4.4 Identidad de Bianchi ContraГӯda

La identidad diferencial de Bianchi $\nabla_{[\lambda}R_{\rho\sigma]\mu\nu} = 0$, al contraer, da:

$$\nabla^\mu G_{\mu\nu} = 0, \qquad G_{\mu\nu} \equiv R_{\mu\nu} - \frac{1}{2}R \, g_{\mu\nu}$$

Esta identidad es **crucial**: garantiza la consistencia de las ecuaciones de Einstein con la conservaciГіn de energГӯa-momento.

### 4.5 Tensor de Weyl

La parte "sin traza" de la curvatura es el **tensor de Weyl** $C_{\rho\sigma\mu\nu}$, que describe la curvatura debida a **campos gravitacionales libres** (mareas, ondas gravitacionales) y no depende de la materia local. En el vacГӯo: $R_{\mu\nu} = 0$ pero $C_{\rho\sigma\mu\nu} \neq 0$.

## вӯҗ 5. Ecuaciones de Campo de Einstein

---

### 5.1 El Camino hacia las Ecuaciones

Einstein buscГі una ecuaciГіn de la forma:

$$\text{GeometrГӯa} = \text{Materia-EnergГӯa}$$

**Lado izquierdo (geometrГӯa):** debe ser un tensor simГ©trico de rango 2, construido solo con la mГ©trica y sus derivadas hasta segundo orden, con divergencia nula. El Гәnico candidato es el **tensor de Einstein**:

$$G_{\mu\nu} \equiv R_{\mu\nu} - \frac{1}{2}R \, g_{\mu\nu}$$

que satisface $\nabla^\mu G_{\mu\nu} = 0$ automГЎticamente (identidad de Bianchi contraГӯda).

**Lado derecho (materia):** el **tensor de energГӯa-momento** $T_{\mu\nu}$, que satisface $\nabla^\mu T_{\mu\nu} = 0$ (conservaciГіn de energГӯa-momento).

### 5.2 Las Ecuaciones de Einstein (1915)

$$\boxed{G_{\mu\nu} + \Lambda g_{\mu\nu} = \frac{8\pi G}{c^4} T_{\mu\nu}}$$

o equivalentemente:

$$R_{\mu\nu} - \frac{1}{2}R \, g_{\mu\nu} + \Lambda g_{\mu\nu} = \frac{8\pi G}{c^4} T_{\mu\nu}$$

donde:
- $R_{\mu\nu}$ es el **tensor de Ricci** (contracciГіn del tensor de Riemann)
- $R = g^{\mu\nu}R_{\mu\nu}$ es el **escalar de curvatura**
- $\Lambda$ es la **constante cosmolГіgica** (introducida por Einstein en 1917)
- $G/c^4 \approx 8.26 \times 10^{-45}$ sВІ/(kgВ·m) es extremadamente pequeГұo вҶ’ ВЎla gravedad es dГ©bil!

### 5.3 Contenido FГӯsico

> *"La materia le dice al espacio-tiempo cГіmo curvarse; el espacio-tiempo le dice a la materia cГіmo moverse."*
> вҖ” John A. Wheeler

- Son **10 ecuaciones diferenciales parciales no lineales** acopladas para $g_{\mu\nu}$
- La **no linealidad** implica que la gravedad se autoacopla (a diferencia de Maxwell)
- En el vacГӯo ($T_{\mu\nu} = 0, \Lambda = 0$): $R_{\mu\nu} = 0$ (ecuaciones de Einstein en el vacГӯo)

### 5.4 LГӯmite Newtoniano

Para campos dГ©biles ($g_{\mu\nu} \approx \eta_{\mu\nu} + h_{\mu\nu}$, $|h| \ll 1$) y velocidades bajas ($v \ll c$), la componente $00$ se reduce a:

$$\nabla^2 \Phi = 4\pi G \rho$$

que es la **ecuaciГіn de Poisson** de la gravedad newtoniana, con $g_{00} \approx -(1 + 2\Phi/c^2)$.

### 5.5 DerivaciГіn desde un Principio de AcciГіn

Las ecuaciones de Einstein se derivan de la **acciГіn de Einstein-Hilbert**:

$$S = \frac{c^4}{16\pi G}\int (R - 2\Lambda)\sqrt{-g} \, d^4x + S_{materia}$$

Variando respecto a $g^{\mu\nu}$: $\delta S = 0 \implies G_{\mu\nu} + \Lambda g_{\mu\nu} = \frac{8\pi G}{c^4}T_{\mu\nu}$

```python
# ============================================================================
# SecciГіn 5b: Ecuaciones de Einstein вҖ” CГЎlculo SimbГіlico con SymPy
# ============================================================================

import sympy as sp
from sympy import symbols, Function, sin, cos, simplify, Matrix, Rational
from sympy.diffgeom import Manifold, Patch, CoordSystem
from IPython.display import display, Math, Markdown

# --- 1. CГЎlculo manual de sГӯmbolos de Christoffel para mГ©trica 2-esfera ---
display(Markdown("### Ejemplo: GeometrГӯa de la 2-esfera $S^2$"))
display(Markdown("MГ©trica: $ds^2 = R^2(d\\theta^2 + \\sin^2\\theta \\, d\\phi^2)$"))

theta, phi_s, R_s = symbols('theta phi R', positive=True)

# MГ©trica
g = Matrix([
    [R_s**2, 0],
    [0, R_s**2 * sin(theta)**2]
])
g_inv = g.inv()

coords = [theta, phi_s]
display(Markdown("**Tensor mГ©trico $g_{\\mu\\nu}$:**"))
display(g)

# SГӯmbolos de Christoffel: О“^О»_ОјОҪ = (1/2) g^О»Пғ (вҲӮ_Ој g_ОҪПғ + вҲӮ_ОҪ g_ОјПғ - вҲӮ_Пғ g_ОјОҪ)
display(Markdown("### SГӯmbolos de Christoffel $\\Gamma^\\lambda_{\\mu\\nu}$"))

Gamma = [[[0 for _ in range(2)] for _ in range(2)] for _ in range(2)]

for lam in range(2):
    for mu in range(2):
        for nu in range(2):
            val = 0
            for sigma in range(2):
                val += Rational(1,2) * g_inv[lam, sigma] * (
                    sp.diff(g[nu, sigma], coords[mu]) +
                    sp.diff(g[mu, sigma], coords[nu]) -
                    sp.diff(g[mu, nu], coords[sigma])
                )
            Gamma[lam][mu][nu] = simplify(val)
            if val != 0:
                display(Math(f"\\Gamma^{{{coords[lam]}}}_{{{coords[mu]}{coords[nu]}}} = {sp.latex(simplify(val))}"))

# Tensor de Riemann: R^ПҒ_ПғОјОҪ
display(Markdown("### Tensor de Riemann $R^\\rho_{\\sigma\\mu\\nu}$"))

Riemann = [[[[0 for _ in range(2)] for _ in range(2)] for _ in range(2)] for _ in range(2)]

for rho in range(2):
    for sigma in range(2):
        for mu in range(2):
            for nu in range(2):
                val = (sp.diff(Gamma[rho][nu][sigma], coords[mu]) 
                     - sp.diff(Gamma[rho][mu][sigma], coords[nu]))
                for lam in range(2):
                    val += (Gamma[rho][mu][lam] * Gamma[lam][nu][sigma] 
                          - Gamma[rho][nu][lam] * Gamma[lam][mu][sigma])
                Riemann[rho][sigma][mu][nu] = simplify(val)
                if simplify(val) != 0:
                    display(Math(f"R^{{{coords[rho]}}}_{{\\ {coords[sigma]}{coords[mu]}{coords[nu]}}} = {sp.latex(simplify(val))}"))

# Tensor de Ricci y escalar de curvatura
display(Markdown("### Tensor de Ricci $R_{\\mu\\nu}$ y Escalar $R$"))

Ricci = Matrix([[0, 0], [0, 0]])
for mu in range(2):
    for nu in range(2):
        val = sum(Riemann[lam][mu][lam][nu] for lam in range(2))
        Ricci[mu, nu] = simplify(val)

display(Markdown("$R_{\\mu\\nu}$:"))
display(Ricci)

# Escalar de Ricci
R_scalar = simplify(sum(g_inv[mu, nu] * Ricci[mu, nu] for mu in range(2) for nu in range(2)))
display(Math(f"R = g^{{\\mu\\nu}} R_{{\\mu\\nu}} = {sp.latex(R_scalar)}"))
display(Markdown(f"**La esfera $S^2$ de radio $R$ tiene curvatura escalar constante $R = 2/R^2$** вң“"))

# --- 2. Ecuaciones de Einstein (simbГіlico) ---
display(Markdown("---"))
display(Markdown("### Ecuaciones de Einstein"))

G_N, c_s, Lambda_s = symbols('G c Lambda')
g_mn, R_mn, T_mn = symbols('g_{\\mu\\nu} R_{\\mu\\nu} T_{\\mu\\nu}')
R_sym = symbols('R')

einstein_eq = sp.Eq(
    sp.Symbol('G_{\\mu\\nu}') + Lambda_s * sp.Symbol('g_{\\mu\\nu}'),
    8 * sp.pi * G_N / c_s**4 * sp.Symbol('T_{\\mu\\nu}')
)
display(Markdown("$$G_{\\mu\\nu} + \\Lambda g_{\\mu\\nu} = \\frac{8\\pi G}{c^4} T_{\\mu\\nu}$$"))
display(Markdown("donde $G_{\\mu\\nu} = R_{\\mu\\nu} - \\frac{1}{2}Rg_{\\mu\\nu}$ es el **tensor de Einstein**."))

# Propiedades
display(Markdown("""
**Propiedades del tensor de Einstein:**
1. SimГ©trico: $G_{\\mu\\nu} = G_{\\nu\\mu}$ вҶ’ 10 ecuaciones independientes
2. Divergencia nula: $\\nabla^\\mu G_{\\mu\\nu} = 0$ (identidad de Bianchi contraГӯda) вҶ’ 4 restricciones  
3. Por tanto: **6 ecuaciones dinГЎmicas** para **10 componentes de $g_{\\mu\\nu}$**  
4. Las 4 componentes restantes corresponden a la **libertad de gauge** (diffeomorfismos)
"""))
```

## рҹ•іпёҸ 6. SoluciГіn de Schwarzschild y Agujeros Negros

---

### 6.1 La Primera SoluciГіn Exacta (1916)

Karl Schwarzschild encontrГі la soluciГіn exacta de las ecuaciones de Einstein en el vacГӯo ($T_{\mu\nu} = 0$) para una masa esfГ©rica, estГЎtica y sin carga:

$$\boxed{ds^2 = -\left(1 - \frac{r_s}{r}\right)c^2 dt^2 + \left(1 - \frac{r_s}{r}\right)^{-1} dr^2 + r^2 \left(d\theta^2 + \sin^2\theta \, d\phi^2\right)}$$

donde el **radio de Schwarzschild** es:

$$r_s = \frac{2GM}{c^2}$$

| Objeto | Masa | $r_s$ |
|--------|------|-------|
| Tierra | $6 \times 10^{24}$ kg | 8.87 mm |
| Sol | $2 \times 10^{30}$ kg | 2.95 km |
| Sgr A* | $4 \times 10^6 M_\odot$ | $1.2 \times 10^{10}$ m |

### 6.2 Estructura del Espacio-Tiempo

- **$r > r_s$**: RegiГіn exterior. La mГ©trica es estГЎtica. Recupera Newton para $r \gg r_s$.
- **$r = r_s$**: **Horizonte de eventos**. Singularidad coordenada (removible cambiando a coordenadas de Eddington-Finkelstein o Kruskal-Szekeres). Los conos de luz se "voltean" aquГӯ.
- **$r < r_s$**: $t$ y $r$ intercambian roles (ВЎ$r$ se vuelve temporal!). Todo cae inevitablemente hacia $r = 0$.
- **$r = 0$**: **Singularidad fГӯsica** genuina ($R_{\mu\nu\rho\sigma}R^{\mu\nu\rho\sigma} \to \infty$).

### 6.3 Г“rbitas GeodГ©sicas

Para una partГӯcula masiva con momento angular $L$ y energГӯa $E$, la ecuaciГіn radial es:

$$\frac{1}{2}\dot{r}^2 + V_{eff}(r) = \frac{1}{2}\left(\frac{E}{mc^2}\right)^2$$

con el **potencial efectivo**:

$$V_{eff}(r) = \frac{1}{2}\left(1 - \frac{r_s}{r}\right)\left(1 + \frac{L^2}{m^2 c^2 r^2}\right)$$

### 6.4 Г“rbitas Circulares y ISCO

- **Г“rbita circular inestable (esfera de fotones):** $r = \frac{3}{2}r_s$
- **ISCO (Innermost Stable Circular Orbit):** $r = 3r_s = 6GM/c^2$
- **PrecesiГіn del perihelio** (por Гіrbita): $\Delta\phi = \frac{6\pi G M}{c^2 a(1-e^2)}$

### 6.5 Corrimiento al Rojo Gravitacional

Un fotГіn emitido a distancia $r_1$ y observado en $r_2 > r_1$:

$$\frac{\lambda_{obs}}{\lambda_{em}} = \sqrt{\frac{1 - r_s/r_2}{1 - r_s/r_1}} > 1$$

La luz se **enrojece** al escapar de un campo gravitacional. Esto fue verificado por el experimento de Pound-Rebka (1959).

```python
# ============================================================================
# SecciГіn 6b: Agujero Negro de Schwarzschild вҖ” VisualizaciГіn 3D y Г“rbitas
# ============================================================================

from scipy.integrate import solve_ivp

# --- 1. Embedding 3D de la geometrГӯa de Schwarzschild ---
# La geometrГӯa del plano ecuatorial (t=const, Оё=ПҖ/2) se puede "incrustar" en 3D
# usando la superficie de Flamm: z(r) = 2вҲҡ(rs(r-rs)) para r вүҘ rs

rs = 2.0  # Radio de Schwarzschild (unidades geomГ©tricas)

fig = plt.figure(figsize=(16, 7))

# Panel 1: Superficie de Flamm (embedding diagram)
ax1 = fig.add_subplot(121, projection='3d')

r_vals = np.linspace(rs * 1.001, 8 * rs, 200)
phi_vals = np.linspace(0, 2 * np.pi, 100)
R, PHI = np.meshgrid(r_vals, phi_vals)

X = R * np.cos(PHI)
Y = R * np.sin(PHI)
Z = 2 * np.sqrt(rs * (R - rs))  # FГіrmula de Flamm

ax1.plot_surface(X, Y, Z, cmap='inferno', alpha=0.7, edgecolor='none')
ax1.plot_surface(X, Y, -Z, cmap='inferno', alpha=0.3, edgecolor='none')  # Puente Einstein-Rosen

# Horizonte de eventos
theta_h = np.linspace(0, 2*np.pi, 100)
ax1.plot(rs * np.cos(theta_h), rs * np.sin(theta_h), np.zeros_like(theta_h), 
         'r-', linewidth=3, label=f'Horizonte $r = r_s = {rs}$')

ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_zlabel('$z$ (embedding)')
ax1.set_title('Diagrama de Flamm\n(GeometrГӯa de Schwarzschild)', fontsize=13)
ax1.legend(fontsize=10)
ax1.view_init(elev=25, azim=45)

# Panel 2: Potencial efectivo y tipos de Гіrbitas
ax2 = fig.add_subplot(122)

r_range = np.linspace(1.5 * rs, 25 * rs, 500)

# V_eff(r) = (1 - rs/r)(1 + LВІ/rВІ) para partГӯculas masivas (en unidades c=G=1)
L_values = [3.5, 4.0, 4.5, 5.0, 6.0]
colors_L = plt.cm.coolwarm(np.linspace(0, 1, len(L_values)))

for i, L in enumerate(L_values):
    V_eff = (1 - rs / r_range) * (1 + L**2 / r_range**2)
    ax2.plot(r_range / rs, V_eff, color=colors_L[i], linewidth=2, label=f'$L/m = {L:.1f}$')

# Fotones: V_eff = (1 - rs/r)/rВІ (sin el tГ©rmino de masa)
V_eff_photon = (1 - rs / r_range) / r_range**2
ax2.plot(r_range / rs, V_eff_photon * 50 + 0.8, 'k--', linewidth=1.5, 
         label='FotГіn (escalado)')

ax2.axhline(y=1, color='gray', linestyle=':', alpha=0.5)
ax2.axvline(x=1, color='red', linestyle='-', alpha=0.3, linewidth=5)
ax2.text(1.05, 0.82, '$r_s$', fontsize=14, color='red')

# ISCO (Innermost Stable Circular Orbit)
ax2.axvline(x=3, color='green', linestyle='--', alpha=0.7)
ax2.text(3.1, 1.08, 'ISCO\n$r = 3r_s$', fontsize=10, color='green')

ax2.set_xlabel('$r / r_s$', fontsize=13)
ax2.set_ylabel('$V_{eff}(r) / (mc^2)$', fontsize=13)
ax2.set_title('Potencial Efectivo de Schwarzschild', fontsize=13)
ax2.legend(fontsize=10, loc='upper right')
ax2.set_xlim(1, 12)
ax2.set_ylim(0.8, 1.1)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# --- 2. Г“rbitas geodГ©sicas (integraciГіn numГ©rica) ---
fig2, ax_orb = plt.subplots(figsize=(10, 10))

def geodesic_rhs(phi, y, L, rs):
    """u'' + u = rs/(2LВІ) + (3/2)rsВ·uВІ  (ecuaciГіn de Binet relativista)"""
    u, du = y
    d2u = -u + rs / (2 * L**2) + 1.5 * rs * u**2
    return [du, d2u]

# Diferentes condiciones iniciales -> diferentes Гіrbitas
orbit_params = [
    (4.5, 1/(6*rs), 0.0, (0, 12*np.pi), 'Г“rbita de precesiГіn', '#2196F3'),
    (4.0, 1/(4*rs), 0.02, (0, 8*np.pi), 'Espiral inestable', '#FF5722'),
    (6.0, 1/(8*rs), 0.0, (0, 10*np.pi), 'Г“rbita cuasi-circular', '#4CAF50'),
    (3.8, 1/(3.5*rs), 0.01, (0, 4*np.pi), 'CaГӯda al horizonte', '#9C27B0'),
]

for L, u0, du0, phi_span, label, color in orbit_params:
    phi_eval = np.linspace(phi_span[0], phi_span[1], 5000)
    
    sol = solve_ivp(geodesic_rhs, phi_span, [u0, du0], t_eval=phi_eval, 
                    args=(L, rs), method='RK45', rtol=1e-10,
                    events=lambda phi, y, L=L, rs=rs: y[0] - 1/rs)  # detener en r = rs
    
    if sol.success and len(sol.t) > 10:
        r_sol = 1.0 / np.maximum(sol.y[0], 1e-6)
        r_sol = np.minimum(r_sol, 30 * rs)  # cortar valores enormes
        x_orb = r_sol * np.cos(sol.t)
        y_orb = r_sol * np.sin(sol.t)
        ax_orb.plot(x_orb, y_orb, color=color, linewidth=1.5, alpha=0.8, label=label)

# Horizonte de eventos
theta_h = np.linspace(0, 2*np.pi, 100)
ax_orb.fill(rs * np.cos(theta_h), rs * np.sin(theta_h), color='black', alpha=0.9)
ax_orb.plot(rs * np.cos(theta_h), rs * np.sin(theta_h), 'r-', linewidth=2)

# ISCO
ax_orb.plot(3*rs * np.cos(theta_h), 3*rs * np.sin(theta_h), 'g--', linewidth=1, alpha=0.5, label='ISCO')

# Esfera de fotones
ax_orb.plot(1.5*rs * np.cos(theta_h), 1.5*rs * np.sin(theta_h), 
            'gold', linewidth=1, linestyle=':', alpha=0.7, label='Esfera de fotones')

ax_orb.set_xlim(-25, 25)
ax_orb.set_ylim(-25, 25)
ax_orb.set_aspect('equal')
ax_orb.set_xlabel('$x / r_s$', fontsize=13)
ax_orb.set_ylabel('$y / r_s$', fontsize=13)
ax_orb.set_title('Г“rbitas GeodГ©sicas en Schwarzschild\n(PrecesiГіn relativista visible)', fontsize=14)
ax_orb.legend(fontsize=10, loc='upper right')
ax_orb.grid(True, alpha=0.2)
plt.tight_layout()
plt.show()

# --- PrecesiГіn del perihelio ---
print("=" * 60)
print("PRECESIГ“N DEL PERIHELIO DE MERCURIO")
print("=" * 60)
G = 6.674e-11
M_sol = 1.989e30
c = 3e8
a_merc = 5.791e10   # semieje mayor (m)
e_merc = 0.2056     # excentricidad
T_merc = 87.969     # periodo orbital (dГӯas)

delta_phi = 6 * np.pi * G * M_sol / (c**2 * a_merc * (1 - e_merc**2))
delta_phi_arcsec_century = delta_phi * (180/np.pi) * 3600 * (36525 / T_merc)

print(f"PrecesiГіn por Гіrbita: {delta_phi*206265:.4f}\"")
print(f"PrecesiГіn por siglo:  {delta_phi_arcsec_century:.2f}\"/siglo")
print(f"Valor observado:      43.11\"/siglo")
print(f"ВЎCoincidencia excelente! Esta fue la primera confirmaciГіn de la RG.")
```

## рҹ”ӯ 7. Predicciones ClГЎsicas: Lente Gravitacional y DeflexiГіn de la Luz

---

### 7.1 GeodГ©sicas Nulas en Schwarzschild

Los fotones siguen geodГ©sicas nulas ($ds^2 = 0$). En la mГ©trica de Schwarzschild:

$$0 = -\left(1 - \frac{r_s}{r}\right)c^2 dt^2 + \left(1 - \frac{r_s}{r}\right)^{-1} dr^2 + r^2 d\phi^2$$

Con las constantes de movimiento $E$ (energГӯa) y $L$ (momento angular), la ecuaciГіn de la Гіrbita del fotГіn es:

$$\left(\frac{du}{d\phi}\right)^2 = \frac{1}{b^2} - u^2 + r_s u^3$$

donde $u = 1/r$ y $b = L/E$ es el **parГЎmetro de impacto**.

### 7.2 DeflexiГіn de la Luz

Para un rayo que pasa a distancia $b \gg r_s$ de una masa $M$, la deflexiГіn total es:

$$\boxed{\alpha = \frac{4GM}{c^2 b} = \frac{2r_s}{b}}$$

Esto es **el doble** de la predicciГіn newtoniana $\alpha_N = 2GM/(c^2 b)$.

**VerificaciГіn histГіrica:** El eclipse solar del 29 de mayo de 1919 (expediciГіn de Eddington) midiГі la deflexiГіn de estrellas cerca del limbo solar: $\alpha = 1.75''$, confirmando la predicciГіn de Einstein.

### 7.3 Lente Gravitacional

Cuando una masa actГәa como "lente" entre una fuente lejana y el observador:

**EcuaciГіn de la lente:**
$$\beta = \theta - \frac{\theta_E^2}{\theta}$$

donde $\theta$ es la posiciГіn angular observada, $\beta$ la posiciГіn real, y:

$$\theta_E = \sqrt{\frac{4GM}{c^2}\frac{D_{LS}}{D_L D_S}}$$

es el **radio del anillo de Einstein**. Cuando $\beta = 0$ (alineaciГіn perfecta), la imagen es un anillo completo.

### 7.4 Tipos de Lentes

- **Lente fuerte:** imГЎgenes mГәltiples, arcos, anillos de Einstein
- **Lente dГ©bil:** distorsiones estadГӯsticas de galaxias de fondo (shear)
- **Microlente:** amplificaciГіn de brillo sin resoluciГіn de imГЎgenes

```python
# ============================================================================
# SecciГіn 7b: Lente Gravitacional вҖ” VisualizaciГіn 3D y SimulaciГіn
# ============================================================================

# --- 1. DeflexiГіn de rayos de luz por una masa puntual ---
fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# Panel 1: Trayectorias de rayos deflectados
ax1 = axes[0]
M_lens = 1.0  # Masa (unidades normalizadas)
rs_lens = 2 * M_lens  # Radio de Schwarzschild

# ParГЎmetros de impacto
b_values = np.linspace(3, 15, 12)

for b in b_values:
    # ГҒngulo de deflexiГіn (aproximaciГіn dГ©bil): Оұ = 4GM/(cВІb) = 2rs/b
    alpha = 2 * rs_lens / b
    
    # Rayo entrante (desde la izquierda)
    x_in = np.linspace(-20, 0, 100)
    y_in = np.full_like(x_in, b)
    
    # Rayo saliente (deflectado)
    x_out = np.linspace(0, 20, 100)
    y_out = b - alpha * x_out  # deflexiГіn lineal
    
    color = plt.cm.plasma(b / 16)
    ax1.plot(x_in, y_in, '-', color=color, linewidth=1, alpha=0.7)
    ax1.plot(x_out, y_out, '-', color=color, linewidth=1, alpha=0.7)
    
    # Rayos simГ©tricos por abajo
    ax1.plot(x_in, -y_in, '-', color=color, linewidth=1, alpha=0.7)
    ax1.plot(x_out, -y_out, '-', color=color, linewidth=1, alpha=0.7)

# Masa (lente)
circle = plt.Circle((0, 0), rs_lens, color='black', zorder=5)
ax1.add_patch(circle)
ax1.text(0, -4, '$M$', fontsize=14, ha='center', color='white',
        bbox=dict(boxstyle='round', facecolor='black'))

# Anillo de Einstein
b_E = np.sqrt(2 * rs_lens * 10)  # para fuente a distancia 10
theta_E = np.linspace(0, 2*np.pi, 100)
ax1.plot(b_E * np.cos(theta_E) * 0.5, b_E * np.sin(theta_E), 
         'gold', linewidth=2, linestyle='--', label=f'Anillo de Einstein ($b_E = {b_E:.1f}$)')

ax1.set_xlim(-20, 20)
ax1.set_ylim(-18, 18)
ax1.set_xlabel('$x$ (unidades de $r_s$)')
ax1.set_ylabel('$y$ (parГЎmetro de impacto $b$)')
ax1.set_title('DeflexiГіn de Rayos de Luz: Lente Gravitacional')
ax1.legend(fontsize=10)
ax1.grid(True, alpha=0.3)
ax1.set_aspect('equal')

# --- Panel 2: DistorsiГіn de una grilla de fondo (efecto lente) ---
ax2 = axes[1]

# Grilla de puntos de fondo
Ngrid = 40
x_bg = np.linspace(-5, 5, Ngrid)
y_bg = np.linspace(-5, 5, Ngrid)
X_bg, Y_bg = np.meshgrid(x_bg, y_bg)

# PosiciГіn de la lente
x_L, y_L = 0, 0
theta_E_val = 1.2  # Radio del anillo de Einstein

# EcuaciГіn de lente: Оё_S = Оё_I - Оё_EВІ/|Оё_I| * Оё_I/|Оё_I|
# Inversa: dado Оё_S, hallar Оё_I (para visualizaciГіn, mapeamos posiciГіn observada)
# Usamos la aproximaciГіn directa: desplazar cada punto
R = np.sqrt((X_bg - x_L)**2 + (Y_bg - y_L)**2)
R = np.maximum(R, 0.1)  # evitar singularidad

# DeflexiГіn
deflection_x = theta_E_val**2 * (X_bg - x_L) / R**2
deflection_y = theta_E_val**2 * (Y_bg - y_L) / R**2

X_lensed = X_bg - deflection_x
Y_lensed = Y_bg - deflection_y

# Dibujar grilla original (tenue)
for i in range(Ngrid):
    ax2.plot(X_bg[i, :], Y_bg[i, :], 'b-', alpha=0.1, linewidth=0.5)
    ax2.plot(X_bg[:, i], Y_bg[:, i], 'b-', alpha=0.1, linewidth=0.5)

# Dibujar grilla deformada
for i in range(Ngrid):
    ax2.plot(X_lensed[i, :], Y_lensed[i, :], 'r-', alpha=0.5, linewidth=0.8)
    ax2.plot(X_lensed[:, i], Y_lensed[:, i], 'r-', alpha=0.5, linewidth=0.8)

# Anillo de Einstein
theta_circ = np.linspace(0, 2*np.pi, 100)
ax2.plot(theta_E_val * np.cos(theta_circ), theta_E_val * np.sin(theta_circ), 
         'gold', linewidth=2.5, linestyle='--', label=f'$\\theta_E = {theta_E_val}$')
ax2.plot(0, 0, 'k*', markersize=15, label='Lente')

ax2.set_xlim(-5, 5)
ax2.set_ylim(-5, 5)
ax2.set_aspect('equal')
ax2.set_title('DistorsiГіn del Fondo por Lente Gravitacional')
ax2.set_xlabel('$x$')
ax2.set_ylabel('$y$')
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.2)

plt.tight_layout()
plt.show()

# --- 3. ГҒngulo de deflexiГіn: verificaciГіn numГ©rica ---
print("=" * 60)
print("DEFLEXIГ“N DE LA LUZ POR EL SOL")
print("=" * 60)
G = 6.674e-11
M_sol = 1.989e30
c = 3e8
R_sol = 6.957e8

alpha_sol = 4 * G * M_sol / (c**2 * R_sol)  # radianes
alpha_arcsec = alpha_sol * 206265  # convertir a segundos de arco

print(f"DeflexiГіn predicha (RG): Оұ = 4GM/(cВІRвҳү) = {alpha_arcsec:.3f}\"")
print(f"DeflexiГіn Newtoniana (ВҪ de RG):             {alpha_arcsec/2:.3f}\"")
print(f"Valor medido por Eddington (1919):          1.75 Вұ 0.06\"")
print(f"\nВЎLa predicciГіn de Einstein fue confirmada!")
```

## рҹҢҠ 8. Ondas Gravitacionales

---

### 8.1 LinealizaciГіn de las Ecuaciones de Einstein

En el **lГӯmite de campo dГ©bil**, escribimos la mГ©trica como una perturbaciГіn del espacio plano:

$$g_{\mu\nu} = \eta_{\mu\nu} + h_{\mu\nu}, \qquad |h_{\mu\nu}| \ll 1$$

Las ecuaciones de Einstein linealizadas, en la **gauge de Lorenz** ($\partial^\mu \bar{h}_{\mu\nu} = 0$, donde $\bar{h}_{\mu\nu} = h_{\mu\nu} - \frac{1}{2}\eta_{\mu\nu}h$):

$$\boxed{\Box \bar{h}_{\mu\nu} = -\frac{16\pi G}{c^4} T_{\mu\nu}}$$

donde $\Box = -\frac{1}{c^2}\partial_t^2 + \nabla^2$ es el d'Alembertiano.

### 8.2 SoluciГіn en el VacГӯo: Ondas Planas

En el vacГӯo ($T_{\mu\nu} = 0$), la ecuaciГіn se reduce a:

$$\Box \bar{h}_{\mu\nu} = 0$$

ВЎEsta es la **ecuaciГіn de onda**! Las soluciones son ondas planas:

$$h_{\mu\nu} = A_{\mu\nu} \cos(k_\alpha x^\alpha) = A_{\mu\nu} \cos(\omega t - \vec{k} \cdot \vec{x})$$

con $k^\mu k_\mu = 0$, es decir, las ondas gravitacionales **viajan a la velocidad de la luz**.

### 8.3 Gauge TT y Polarizaciones

En la **gauge transversa sin traza** (TT), para una onda propagГЎndose en $z$:

$$h_{\mu\nu}^{TT} = \begin{pmatrix} 0 & 0 & 0 & 0 \\ 0 & h_+ & h_\times & 0 \\ 0 & h_\times & -h_+ & 0 \\ 0 & 0 & 0 & 0 \end{pmatrix} \cos(\omega t - kz)$$

Hay exactamente **dos polarizaciones independientes**:
- $h_+$ (**plus**): estira/comprime en $x$ y $y$ alternadamente
- $h_\times$ (**cross**): lo mismo, rotado 45В°

### 8.4 FГіrmula del Cuadrupolo

La potencia radiada por una fuente de ondas gravitacionales es:

$$\boxed{P = \frac{G}{5c^5} \left\langle \dddot{Q}_{ij} \dddot{Q}^{ij} \right\rangle}$$

donde $Q_{ij} = \int \rho\left(x_i x_j - \frac{1}{3}\delta_{ij}r^2\right) d^3x$ es el **momento cuadrupolar de masa** (reducido).

NГіtese el factor $G/c^5 \approx 2.6 \times 10^{-53}$ W$^{-1}$: la gravedad es extremadamente dГ©bil como radiadora.

### 8.5 DetecciГіn: LIGO y la Primera ObservaciГіn (2015)

El 14 de septiembre de 2015, los detectores LIGO observaron la seГұal GW150914: la fusiГіn de dos agujeros negros de $\sim 36 M_\odot$ y $\sim 29 M_\odot$, produciendo un agujero negro de $\sim 62 M_\odot$. La energГӯa radiada: $\sim 3 M_\odot c^2$ en ondas gravitacionales en $\sim 0.2$ s.

La amplitud de strain medida: $h \sim 10^{-21}$, correspondiente a un cambio de longitud $\Delta L \sim 10^{-18}$ m (ВЎuna milГ©sima del diГЎmetro de un protГіn!) en los brazos de 4 km.

## рҹҢҢ 9. CosmologГӯa Relativista: MГ©trica de Friedmann-Robertson-Walker

---

### 9.1 El Principio CosmolГіgico

A escalas suficientemente grandes ($\gtrsim 100$ Mpc), el universo es:
- **HomogГ©neo:** las propiedades son las mismas en todos los puntos
- **IsГіtropo:** las propiedades son las mismas en todas las direcciones

### 9.2 La MГ©trica FRW

La mГ©trica mГЎs general compatible con homogeneidad e isotropГӯa es:

$$\boxed{ds^2 = -c^2 dt^2 + a^2(t)\left[\frac{dr^2}{1-kr^2} + r^2\left(d\theta^2 + \sin^2\theta \, d\phi^2\right)\right]}$$

donde:
- $a(t)$ es el **factor de escala** (la "distancia" entre galaxias crece como $a(t)$)
- $k = +1, 0, -1$ determina la curvatura espacial (cerrado, plano, abierto)
- Las coordenadas $(r, \theta, \phi)$ son **comГіviles** (se mueven con la expansiГіn)

### 9.3 Ecuaciones de Friedmann

Insertando la mГ©trica FRW en las ecuaciones de Einstein con un fluido perfecto $T^{\mu\nu} = (\rho + p/c^2)u^\mu u^\nu + p g^{\mu\nu}$:

**Primera ecuaciГіn de Friedmann** (componente $00$):

$$\boxed{\left(\frac{\dot{a}}{a}\right)^2 = \frac{8\pi G}{3}\rho - \frac{kc^2}{a^2} + \frac{\Lambda c^2}{3}}$$

**Segunda ecuaciГіn de Friedmann** (ecuaciГіn de aceleraciГіn):

$$\boxed{\frac{\ddot{a}}{a} = -\frac{4\pi G}{3}\left(\rho + \frac{3p}{c^2}\right) + \frac{\Lambda c^2}{3}}$$

### 9.4 ParГЎmetro de Hubble y ParГЎmetros de Densidad

El **parГЎmetro de Hubble** es $H(t) = \dot{a}/a$, con valor actual $H_0 \approx 67.4$ km/s/Mpc.

Definiendo la **densidad crГӯtica** $\rho_c = 3H^2/(8\pi G)$ y los parГЎmetros:

$$\Omega_m = \frac{\rho_m}{\rho_c}, \quad \Omega_r = \frac{\rho_r}{\rho_c}, \quad \Omega_\Lambda = \frac{\Lambda c^2}{3H^2}, \quad \Omega_k = -\frac{kc^2}{a^2 H^2}$$

La primera ecuaciГіn de Friedmann se reescribe elegantemente:

$$\boxed{\Omega_m + \Omega_r + \Omega_\Lambda + \Omega_k = 1}$$

### 9.5 EvoluciГіn de los Componentes

Para un fluido con ecuaciГіn de estado $p = w\rho c^2$:
- **Materia** ($w = 0$): $\rho_m \propto a^{-3}$ вҖ” diluciГіn por volumen
- **RadiaciГіn** ($w = 1/3$): $\rho_r \propto a^{-4}$ вҖ” diluciГіn + corrimiento al rojo
- **EnergГӯa oscura** ($w = -1$): $\rho_\Lambda = \text{const}$ вҖ” densidad constante

### 9.6 Nuestro Universo: ОӣCDM

Las observaciones (CMB, supernovas Ia, BAO) indican:

$$\Omega_m \approx 0.315, \quad \Omega_\Lambda \approx 0.685, \quad \Omega_k \approx 0 \quad \text{(plano)}$$

El universo estГЎ **acelerando** su expansiГіn ($\ddot{a} > 0$) debido a la energГӯa oscura.

```python
# ============================================================================
# SecciГіn 9b: CosmologГӯa FRW вҖ” EvoluciГіn del Factor de Escala (AnimaciГіn)
# ============================================================================

from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# --- EcuaciГіn de Friedmann: (И§/a)ВІ = HвӮҖВІ[О©r/aвҒҙ + О©m/aВі + О©k/aВІ + О©Оӣ] ---
# IntegraciГіn numГ©rica para diferentes composiciones

from scipy.integrate import solve_ivp

H0 = 67.4  # km/s/Mpc (Planck 2018)
H0_si = H0 * 1e3 / (3.086e22)  # en sвҒ»В№

# Modelos cosmolГіgicos: (О©m, О©Оӣ, nombre, color)
modelos = [
    (0.3, 0.7, 'ОӣCDM (nuestro universo)', '#2196F3'),
    (1.0, 0.0, 'Einstein-de Sitter ($\\Omega_m=1$)', '#FF5722'),
    (0.3, 0.0, 'Universo abierto ($\\Omega_\\Lambda=0$)', '#4CAF50'),
    (0.0, 1.0, 'de Sitter ($\\Omega_\\Lambda=1$)', '#9C27B0'),
    (4.0, 0.0, 'Universo cerrado ($\\Omega_m=4$)', '#FF9800'),
]

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# --- Panel 1: Factor de escala a(t) ---
ax1 = axes[0]

for Om, OL, nombre, color in modelos:
    Ok = 1 - Om - OL
    
    def friedmann(t, y):
        a = y[0]
        if a <= 0:
            return [0]
        H2 = H0_si**2 * (Om / a**3 + Ok / a**2 + OL)
        if H2 < 0:
            return [0]
        return [a * np.sqrt(H2)]
    
    t_span = (0, 5e17)  # ~15.8 Gyr en segundos
    t_eval = np.linspace(1e10, 5e17, 2000)
    
    sol = solve_ivp(friedmann, t_span, [1e-6], t_eval=t_eval, 
                    method='RK45', max_step=1e15, rtol=1e-8)
    
    if sol.success:
        t_Gyr = sol.t / (3.156e16)  # segundos a Gyr
        a_vals = sol.y[0]
        # Normalizar: a(ahora) вүҲ 1
        a_vals = a_vals / a_vals[-1] if a_vals[-1] > 0 else a_vals
        ax1.plot(t_Gyr, a_vals, color=color, linewidth=2.5, label=nombre)

ax1.set_xlabel('Tiempo (Gyr)', fontsize=13)
ax1.set_ylabel('Factor de escala $a(t)$', fontsize=13)
ax1.set_title('EvoluciГіn del Universo: EcuaciГіn de Friedmann', fontsize=14)
ax1.legend(fontsize=10, loc='upper left')
ax1.set_xlim(0, 16)
ax1.set_ylim(0, 3)
ax1.axvline(x=13.8, color='gray', linestyle=':', alpha=0.7)
ax1.text(13.9, 0.1, 'Hoy\n13.8 Gyr', fontsize=10, color='gray')
ax1.grid(True, alpha=0.3)

# --- Panel 2: Diagrama О©m vs О©Оӣ ---
ax2 = axes[1]
Om_range = np.linspace(0, 2, 300)
OL_range = np.linspace(-0.5, 2, 300)
Om_grid, OL_grid = np.meshgrid(Om_range, OL_range)

# Regiones
# Plano: О©m + О©Оӣ = 1
ax2.plot(Om_range, 1 - Om_range, 'k-', linewidth=2, label='Plano ($k=0$)')
ax2.fill_between(Om_range, 1 - Om_range, 2, alpha=0.1, color='blue')
ax2.fill_between(Om_range, -0.5, 1 - Om_range, alpha=0.1, color='red')
ax2.text(0.1, 1.5, 'Abierto ($k<0$)', fontsize=12, color='blue')
ax2.text(1.2, 0.1, 'Cerrado ($k>0$)', fontsize=12, color='red')

# LГӯnea de aceleraciГіn/desaceleraciГіn: О©Оӣ = О©m/2
ax2.plot(Om_range, Om_range/2, 'g--', linewidth=2, label='$q_0 = 0$ (aceleraciГіn)')
ax2.text(0.3, 0.5, 'Acelerando', fontsize=11, color='green', rotation=25)
ax2.text(1.0, 0.3, 'Desacelerando', fontsize=11, color='darkred', rotation=25)

# Punto observado (Planck 2018)
ax2.plot(0.315, 0.685, '*', color='gold', markersize=20, markeredgecolor='black', 
         markeredgewidth=1.5, zorder=5, label='Planck 2018')

# Einstein estГЎtico
ax2.plot(1.0, 1.0, 'D', color='purple', markersize=10, label='Einstein estГЎtico')

ax2.set_xlabel('$\\Omega_m$', fontsize=14)
ax2.set_ylabel('$\\Omega_\\Lambda$', fontsize=14)
ax2.set_title('Espacio de ParГЎmetros CosmolГіgicos', fontsize=14)
ax2.legend(fontsize=10, loc='lower right')
ax2.set_xlim(0, 2)
ax2.set_ylim(-0.5, 2)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# --- 3. AnimaciГіn: ExpansiГіn del universo ---
fig_anim, ax_anim = plt.subplots(figsize=(8, 8))

np.random.seed(42)
N_galaxias = 80
theta_g = np.random.uniform(0, 2*np.pi, N_galaxias)
r_g = np.random.uniform(0.1, 1.0, N_galaxias)
x_g = r_g * np.cos(theta_g)
y_g = r_g * np.sin(theta_g)
colors_g = plt.cm.plasma(np.random.uniform(0.2, 0.9, N_galaxias))

def animate_expansion(frame):
    ax_anim.clear()
    a = 0.3 + frame * 0.02  # factor de escala creciente
    ax_anim.scatter(a * x_g, a * y_g, c=colors_g, s=30 + 20*r_g, alpha=0.8, edgecolors='white', linewidth=0.3)
    
    # Dibujar flechas de Hubble (velocidad proporcional a distancia)
    if frame > 10:
        for i in range(0, N_galaxias, 8):
            dx = 0.03 * a * x_g[i]
            dy = 0.03 * a * y_g[i]
            if abs(dx) + abs(dy) > 0.005:
                ax_anim.arrow(a*x_g[i], a*y_g[i], dx, dy, 
                            head_width=0.02, head_length=0.01, fc='red', ec='red', alpha=0.6)
    
    ax_anim.set_xlim(-2, 2)
    ax_anim.set_ylim(-2, 2)
    ax_anim.set_aspect('equal')
    t_gyr = 0.5 + frame * 0.25
    ax_anim.set_title(f'ExpansiГіn del Universo вҖ” $a = {a:.2f}$, $t \\approx {t_gyr:.1f}$ Gyr', fontsize=14)
    ax_anim.set_xlabel('x (unidades comГіviles escaladas)')
    ax_anim.set_ylabel('y (unidades comГіviles escaladas)')
    ax_anim.grid(True, alpha=0.2)
    return []

anim_expansion = FuncAnimation(fig_anim, animate_expansion, frames=60, interval=100, blit=True)
plt.close(fig_anim)
HTML(anim_expansion.to_jshtml())
```
