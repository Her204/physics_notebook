## 1. Introducción y Motivación Histórica

### El nacimiento de la teoría de juegos

La **teoría de juegos** es la rama de la matemática que estudia situaciones de **interacción estratégica**: escenarios donde el resultado para cada participante depende no solo de sus propias decisiones, sino también de las decisiones de los demás.

#### Hitos fundamentales:

| Año | Contribución | Autor(es) |
|------|-------------|-----------|
| 1713 | Primera solución de estrategia mixta (Le Her) | James Waldegrave |
| 1838 | Modelo de competencia en cantidades (duopolio) | Antoine Augustin Cournot |
| 1913 | Primer teorema formal de juegos (forma extensiva) | Ernst Zermelo |
| 1928 | Teorema minimax para juegos de suma cero | John von Neumann |
| 1944 | *Theory of Games and Economic Behavior* | von Neumann & Morgenstern |
| 1950 | Equilibrio de Nash (juegos no cooperativos) | John F. Nash Jr. |
| 1953 | Valor de Shapley (juegos cooperativos) | Lloyd Shapley |
| 1967-68 | Juegos bayesianos (información incompleta) | John Harsanyi |
| 1973 | Estrategia evolutivamente estable | John Maynard Smith & George Price |
| 1975 | Equilibrio perfecto en subjuegos | Reinhard Selten |
| 1981 | Diseño de mecanismos moderno | Roger Myerson |
| 2012 | Teoría de emparejamiento estable (Nobel) | Alvin Roth & Lloyd Shapley |
| 2020 | Subastas mejoradas (Nobel) | Paul Milgrom & Robert Wilson |

### ¿Por qué es fundamental?

La teoría de juegos unifica fenómenos aparentemente dispares:

- **Economía:** competencia de mercado, subastas, negociación salarial
- **Biología:** evolución de comportamiento altruista, competencia reproductiva
- **Física:** gases en equilibrio termodinámico, redes complejas, teoría cuántica de la información
- **Sociología:** normas sociales, cooperación, conflicto institucional
- **Ciencias de la computación:** diseño de protocolos, criptografía, aprendizaje multiagente

El poder unificador reside en que la estructura matemática subyacente — **optimización bajo incertidumbre estratégica** — es la misma independientemente del dominio.

## 2. Fundamentos Formales

### 2.1 Definición: Juego en forma normal (o estratégica)

Un **juego en forma normal** es una tupla $\Gamma = (N, (S_i)_{i \in N}, (u_i)_{i \in N})$ donde:

- $N = \{1, 2, \ldots, n\}$ es el conjunto finito de **jugadores**
- $S_i$ es el conjunto de **estrategias puras** del jugador $i$
- $u_i : S_1 \times S_2 \times \cdots \times S_n \to \mathbb{R}$ es la función de **utilidad** (pago) del jugador $i$

El **perfil de estrategias** $s = (s_1, s_2, \ldots, s_n) \in S = \prod_{i \in N} S_i$ es un vector que especifica la estrategia elegida por cada jugador.

**Notación fundamental:** Usamos $s_{-i} = (s_1, \ldots, s_{i-1}, s_{i+1}, \ldots, s_n)$ para denotar las estrategias de todos los jugadores *excepto* $i$. Así podemos escribir $s = (s_i, s_{-i})$.

### 2.2 Estrategias mixtas

Una **estrategia mixta** del jugador $i$ es una distribución de probabilidad sobre $S_i$:

$$\sigma_i \in \Delta(S_i) = \left\{ p \in \mathbb{R}^{|S_i|}_{\geq 0} : \sum_{s_i \in S_i} p(s_i) = 1 \right\}$$

El **simplex** $\Delta(S_i)$ es un objeto geométrico fundamental: para $|S_i| = 2$ es un segmento, para $|S_i| = 3$ es un triángulo, etc.

La utilidad esperada bajo estrategias mixtas se extiende por linealidad:

$$u_i(\sigma_1, \ldots, \sigma_n) = \sum_{s \in S} \left(\prod_{j \in N} \sigma_j(s_j)\right) u_i(s)$$

### 2.3 Conceptos de dominancia

**Definición (Dominancia estricta):** La estrategia $s_i$ es **estrictamente dominada** por $s_i'$ si:

$$u_i(s_i', s_{-i}) > u_i(s_i, s_{-i}) \quad \forall s_{-i} \in S_{-i}$$

**Definición (Dominancia débil):** $s_i$ es **débilmente dominada** por $s_i'$ si:

$$u_i(s_i', s_{-i}) \geq u_i(s_i, s_{-i}) \quad \forall s_{-i} \in S_{-i}$$

con desigualdad estricta para al menos un $s_{-i}$.

**Principio de racionalidad:** Un jugador racional nunca jugará una estrategia estrictamente dominada.

**IESDS** (Eliminación Iterada de Estrategias Estrictamente Dominadas): proceso de eliminar sucesivamente estrategias dominadas. El orden de eliminación no importa (el resultado es único para dominancia estricta, pero puede depender del orden para dominancia débil).

### 2.4 Mejor respuesta

La **correspondencia de mejor respuesta** del jugador $i$ dado $s_{-i}$ es:

$$BR_i(s_{-i}) = \arg\max_{s_i \in S_i} u_i(s_i, s_{-i})$$

Para estrategias mixtas:

$$BR_i(\sigma_{-i}) = \arg\max_{\sigma_i \in \Delta(S_i)} u_i(\sigma_i, \sigma_{-i})$$

**Lema fundamental:** $\sigma_i^*$ es mejor respuesta a $\sigma_{-i}$ si y solo si toda estrategia pura en el **soporte** de $\sigma_i^*$ es mejor respuesta a $\sigma_{-i}$:

$$\text{supp}(\sigma_i^*) \subseteq BR_i(\sigma_{-i})$$

Este lema es la clave computacional para encontrar equilibrios de Nash en estrategias mixtas.

```python
# --- Visualización: El simplex de estrategias mixtas para 3 estrategias ---

fig = plt.figure(figsize=(12, 5))

# Simplex 2D (3 estrategias)
ax1 = fig.add_subplot(121)

# Triángulo del simplex
triangle = plt.Polygon([[0, 0], [1, 0], [0.5, np.sqrt(3)/2]], 
                         fill=True, alpha=0.15, color='steelblue', edgecolor='navy', linewidth=2)
ax1.add_patch(triangle)

# Vértices = estrategias puras
vertices = np.array([[0, 0], [1, 0], [0.5, np.sqrt(3)/2]])
labels = ['$s_1$ (1,0,0)', '$s_2$ (0,1,0)', '$s_3$ (0,0,1)']
for v, l in zip(vertices, labels):
    ax1.plot(*v, 'o', markersize=10, color='navy')
    offset = v - np.array([0.5, np.sqrt(3)/6])
    ax1.annotate(l, xy=v, xytext=v + 0.08*offset/np.linalg.norm(offset+1e-10), fontsize=11, ha='center')

# Punto ejemplo de estrategia mixta
p = np.array([0.5, 0.3, 0.2])
point = p[0]*vertices[0] + p[1]*vertices[1] + p[2]*vertices[2]
ax1.plot(*point, 's', markersize=12, color='crimson', zorder=5)
ax1.annotate(f'$\\sigma = ({p[0]}, {p[1]}, {p[2]})$', xy=point, 
             xytext=point + np.array([0.12, 0.05]), fontsize=11, color='crimson',
             arrowprops=dict(arrowstyle='->', color='crimson'))

ax1.set_xlim(-0.15, 1.15)
ax1.set_ylim(-0.15, 1.05)
ax1.set_aspect('equal')
ax1.set_title('Simplex $\\Delta(S_i)$ para 3 estrategias', fontsize=13)
ax1.axis('off')

# Simplex 1D (2 estrategias) - segmento
ax2 = fig.add_subplot(122)
ax2.plot([0, 1], [0, 0], 'navy', linewidth=3)
ax2.plot(0, 0, 'o', markersize=12, color='navy')
ax2.plot(1, 0, 'o', markersize=12, color='navy')
ax2.annotate('$s_1$\n$p=1$', xy=(0, 0), xytext=(0, -0.06), ha='center', fontsize=11)
ax2.annotate('$s_2$\n$p=0$', xy=(1, 0), xytext=(1, -0.06), ha='center', fontsize=11)

p_example = 0.7
ax2.plot(1-p_example, 0, 's', markersize=12, color='crimson', zorder=5)
ax2.annotate(f'$\\sigma = ({p_example}, {1-p_example})$', xy=(1-p_example, 0), 
             xytext=(1-p_example, 0.04), fontsize=11, color='crimson', ha='center')

ax2.set_xlim(-0.15, 1.15)
ax2.set_ylim(-0.12, 0.12)
ax2.set_aspect('equal')
ax2.set_title('Simplex $\\Delta(S_i)$ para 2 estrategias', fontsize=13)
ax2.axis('off')

plt.suptitle('Espacio de estrategias mixtas como simplex', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

## 3. Juegos en Forma Normal y Equilibrio de Nash

### 3.1 Juegos clásicos fundamentales

#### Dilema del Prisionero

Dos sospechosos son interrogados por separado. Cada uno puede **cooperar** (C) con el otro (callar) o **desertar** (D, delatar):

|  | C | D |
|--|---|---|
| **C** | (-1, -1) | (-3, 0) |
| **D** | (0, -3) | (-2, -2) |

**Observación clave:** (D, D) es el único equilibrio de Nash, pero (C, C) es Pareto-superior. Esta tensión entre racionalidad individual y bienestar colectivo es el corazón de muchos problemas en economía, sociología y biología.

#### Batalla de los Sexos (Coordinación)

|  | Ópera | Fútbol |
|--|-------|--------|
| **Ópera** | (3, 2) | (0, 0) |
| **Fútbol** | (0, 0) | (2, 3) |

Dos equilibrios de Nash en estrategias puras: (Ópera, Ópera) y (Fútbol, Fútbol). Problema de **coordinación** con preferencias asimétricas.

#### Halcón-Paloma (Chicken)

|  | Paloma | Halcón |
|--|--------|--------|
| **Paloma** | (3, 3) | (1, 5) |
| **Halcón** | (5, 1) | (0, 0) |

Dos equilibrios puros asimétricos. Modelo fundamental en biología evolutiva y conflicto internacional.

### 3.2 Equilibrio de Nash

**Definición (Nash, 1950):** Un perfil de estrategias $s^* = (s_1^*, \ldots, s_n^*)$ es un **equilibrio de Nash** si ningún jugador puede mejorar unilateralmente su pago:

$$u_i(s_i^*, s_{-i}^*) \geq u_i(s_i, s_{-i}^*) \quad \forall s_i \in S_i, \quad \forall i \in N$$

Equivalentemente, $s^*$ es un equilibrio de Nash si y solo si:

$$s_i^* \in BR_i(s_{-i}^*) \quad \forall i \in N$$

Es decir, **cada jugador está jugando una mejor respuesta a las estrategias de los demás**.

### 3.3 Propiedades fundamentales del equilibrio de Nash

1. **Todo equilibrio de Nash sobrevive a IESDS** (pero no al revés)
2. **Puede no existir en estrategias puras** (ej: Matching Pennies)
3. **Puede no ser único** (ej: Batalla de los Sexos)
4. **Puede no ser eficiente** (ej: Dilema del Prisionero)
5. **Es un punto fijo** de la correspondencia de mejor respuesta

```python
# --- Análisis computacional de juegos clásicos ---

def crear_juego_bimatrix(A, B, nombres_s1, nombres_s2, titulo):
    """Analiza un juego bimatricial: encuentra equilibrios de Nash en estrategias puras."""
    n1, n2 = A.shape
    equilibrios = []
    
    for i in range(n1):
        for j in range(n2):
            # Verificar si i es mejor respuesta a j (jugador 1)
            es_br1 = A[i, j] == np.max(A[:, j])
            # Verificar si j es mejor respuesta a i (jugador 2)
            es_br2 = B[i, j] == np.max(B[i, :])
            if es_br1 and es_br2:
                equilibrios.append((i, j))
    
    return equilibrios

def visualizar_juego(A, B, nombres_s1, nombres_s2, titulo, ax):
    """Visualiza la matriz de pagos con equilibrios de Nash marcados."""
    n1, n2 = A.shape
    equilibrios = crear_juego_bimatrix(A, B, nombres_s1, nombres_s2, titulo)
    
    # Crear tabla visual
    cell_text = []
    cell_colors = []
    for i in range(n1):
        row_text = []
        row_colors = []
        for j in range(n2):
            row_text.append(f'({A[i,j]:.0f}, {B[i,j]:.0f})')
            if (i, j) in equilibrios:
                row_colors.append('#FFD700')  # Dorado para Nash
            else:
                row_colors.append('#E8E8E8')
        cell_text.append(row_text)
        cell_colors.append(row_colors)
    
    ax.axis('off')
    tabla = ax.table(cellText=cell_text, rowLabels=nombres_s1, colLabels=nombres_s2,
                     cellColours=cell_colors, loc='center', cellLoc='center')
    tabla.auto_set_font_size(False)
    tabla.set_fontsize(13)
    tabla.scale(1.5, 2.0)
    
    eq_str = ', '.join([f'({nombres_s1[i]}, {nombres_s2[j]})' for i, j in equilibrios])
    ax.set_title(f'{titulo}\nNash (puro): {eq_str if eq_str else "∅"}', fontsize=13, fontweight='bold', pad=20)

# --- Juegos clásicos ---
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Dilema del Prisionero
A_pd = np.array([[-1, -3], [0, -2]])
B_pd = np.array([[-1, 0], [-3, -2]])
visualizar_juego(A_pd, B_pd, ['C', 'D'], ['C', 'D'], 'Dilema del Prisionero', axes[0])

# Batalla de los Sexos
A_bos = np.array([[3, 0], [0, 2]])
B_bos = np.array([[2, 0], [0, 3]])
visualizar_juego(A_bos, B_bos, ['Ópera', 'Fútbol'], ['Ópera', 'Fútbol'], 'Batalla de los Sexos', axes[1])

# Halcón-Paloma
A_hd = np.array([[3, 1], [5, 0]])
B_hd = np.array([[3, 5], [1, 0]])
visualizar_juego(A_hd, B_hd, ['Paloma', 'Halcón'], ['Paloma', 'Halcón'], 'Halcón-Paloma', axes[2])

plt.suptitle('Juegos Clásicos: Equilibrios de Nash en Estrategias Puras (dorado)', 
             fontsize=15, fontweight='bold', y=1.05)
plt.tight_layout()
plt.show()

print("Nota: El dorado indica casillas que son equilibrios de Nash en estrategias puras.")
```
